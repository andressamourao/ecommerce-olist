## MERGE THE DATA

```python
# merge all the data referring to the customer

# it has the order_id as a common point
customer_data = pd.merge(orders, order_reviews, on="order_id")
customer_data = pd.merge(customer_data, order_payments, on="order_id")

# check if it works
customer_data.head()
```

```python
# to the customer merge (customer_data) add the customers dataset, with the customer_id as a common point
customer_data = pd.merge(customer_data, customers, on="customer_id")

costumer_data.head()
```

```python
# merging all datasets referring to the products on product_id
products_data = pd.merge(order_items, products, on="product_id")
```

```python
# merging all datasets referring to the sellers on seller_id
sellers_data = pd.merge(products_data, sellers, on="seller_id")
sellers_data.head()
```

```python
# add the translation of the categories to the dataset sellers_data on product_category_name
sellers_data = pd.merge(sellers_data, product_translation, on="product_category_name")

sellers_data.head()
```

```python
# merge the customer dataset with the sellers dataset on order_id

df_ecommerce = pd.merge(customer_data, sellers_data, on="order_id")
df_ecommerce.head()
```

## Cleaning the data

```python
df_ecommerce.isnull().sum()
```

```python
# considering we have a large amount of null data in the order_delivered_customer_date, the best option is to remove these lines
df_ecommerce.dropna(how="any", inplace=True) # remove ANY lines with null values, in place to record the modification
```

```python
df_ecommerce.shape
```

```python
# Removing the timestamp and converting the date format
```

```python
df_ecommerce["order_purchase_timestamp"] = pd.to_datetime(df_ecommerce["order_purchase_timestamp"]).dt.date 
df_ecommerce["order_estimated_delivery_date"] = pd.to_datetime(df_ecommerce["order_estimated_delivery_date"]).dt.date
df_ecommerce["order_delivered_customer_date"] = pd.to_datetime(df_ecommerce["order_delivered_customer_date"]).dt.date
df_ecommerce["shipping_limit_date"] = pd.to_datetime(df_ecommerce["shipping_limit_date"]).dt.date
```

```python
# How many days supposedly should take to product arrive?
df_ecommerce["estimated_days"] = df_ecommerce["order_estimated_delivery_date"] - df_ecommerce["order_purchase_timestamp"]
```

```python
df_ecommerce.shape
```

```python
# How many actual days it took for the arriving of the product?
df_ecommerce["delivery_days"] = df_ecommerce["order_delivered_customer_date"] - df_ecommerce["order_purchase_timestamp"]
```

```python
# After purchasing the product, how many days does it take to be shipped?
df_ecommerce['shipping_days'] = df_ecommerce["shipping_limit_date"] - df_ecommerce["order_purchase_timestamp"]
```

```python
# Transforming the timedelta to get a interger

cols = ["estimated_days","delivery_days","shipping_days"]

for label in cols:
    df_ecommerce[label] = df_ecommerce[label].apply(lambda x: x.days)
```
